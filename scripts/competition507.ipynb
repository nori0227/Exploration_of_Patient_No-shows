## Regression: Continuous Outcome Variable 

```python
import numpy as np
import pandas as pd
from supervised.automl import AutoML

df = pd.read_csv('/content/output_noshow1_balanced_2022-11-16.csv')
x_columns = [c for c in df.columns if c != "patient_age"]
X = df[x_columns]
y = df["patient_age"]
y
```

```python
x_columns
```

```python
X
```

```python
y
```

```python
automl = AutoML()
automl.fit(X, y)
```

```python
automl.report()
```

```python
df["noshow_predictions"] = automl.predict(X)
```

```python
print("noshow_predictions")
print(df[["patient_age", "noshow_predictions"]].head())
```
